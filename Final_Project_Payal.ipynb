```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py

```

```python
data = pd.read_csv('diabetes.csv')
```

```python
fig, axs = plt.subplots(3, 3, figsize=(20, 20))
axs[0, 0].plot(data['Pregnancies'], c='r')
axs[0, 0].set_xlabel('Outcome')
axs[0, 0].set_ylabel('Pregnancies')
axs[0, 0].set_title('Pregnancies')

axs[0, 1].plot(data['Glucose'], c='g')
axs[0, 1].set_xlabel('Outcome')
axs[0, 1].set_ylabel('Glucose')
axs[0, 1].set_title('Glucose')


axs[0, 2].plot(data['BloodPressure'], c='b')
axs[0, 2].set_xlabel('Outcome')
axs[0, 2].set_ylabel('BloodPressure')
axs[0, 2].set_title('BloodPressure')

axs[1, 0].plot(data['SkinThickness'], c='y')
axs[1, 0].set_xlabel('Outcome')
axs[1, 0].set_ylabel('SkinThickness')
axs[1, 0].set_title('SkinThickness')

axs[1, 1].plot(data['Insulin'], c='m')
axs[1, 1].set_xlabel('Outcome')
axs[1, 1].set_ylabel('Insulin')
axs[1, 1].set_title('Insulin')

axs[1, 2].plot(data['BMI'], c='c')
axs[1, 2].set_xlabel('Outcome')
axs[1, 2].set_ylabel('BMI')
axs[1, 2].set_title('BMI')

axs[2, 0].plot(data['DiabetesPedigreeFunction'], c='k')
axs[2, 0].set_xlabel('Outcome')
axs[2, 0].set_ylabel('DiabetesPedigreeFunction')
axs[2, 0].set_title('DiabetesPedigreeFunction')

axs[2, 1].plot(data['Age'], c='orange')
axs[2, 1].set_xlabel('Outcome')
axs[2, 1].set_ylabel('Age')
```

The plots imply that the data is varied and spread out over a range and hence this is not the best way to represnt the correlation between the predictor variables and the outcome.

```python
# @title Scatterplots for each Predictor Variable:
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Pregnancies'], y=data['Outcome'], color='#0f5a3e')
plt.xlabel('Pregnancies')
plt.ylabel('Outcome')
plt.title('Pregnancies vs Outcome', fontsize=16, color='Red')
plt.show()
```

This scatterplot implies that there is an equal probability of being incident with diabetes if the number of pregnancies is between 0-13. However mmore than 13 pregnancies shows an outcome of 1 which implies being diagnosed with diabetes. Hence the higher the number of pregnancies, the higher probability of being incident with diabetes.

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=data['Glucose'], y=data['Outcome'], color='#0f5a3e')
plt.xlabel('Glucose')
plt.ylabel('Outcome')
plt.title('Glucose vs Outcome', fontsize=16, color='Red')
plt.show()
```

This scatterplot shows that the spread of values is more at 0 but the higher the glucose level goes, the probability of being diabetes does not become 0. It is always 1. For some values there is an equal chance but for some values the result is positive. Hence this concludes a directly proportional between glucose levels and the probability of being diagnosed with diabetes.
