```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
data = pd.read_csv('nutrition-drinks.csv')
```

```python
plt.figure(figsize=(15,15))
plt.plot(data['Calories'], data['Fat (g)'], data['Protein'], data['Carb. (g)'])

plt.title(' Calories vs Fat (g) vs Protein vs (g) Carb. (g) Line Plot', fontsize=20, color='red')

```

```python
plt.subplot(1,2,1)

plt.plot(data['Calories'], data['Fat (g)'], marker='o', color='blue', markersize='0.2', markerfacecolor='red', linestyle='-')

plt.plot(data['Calories'], data['Fat (g)'], marker='o', color='blue', markersize='0.2', markerfacecolor='red', linestyle='-')
plt.xlabel('Calories')
plt.ylabel('Fat (g)')
plt.title('Calories vs Fat (g) Line Plot', fontsize=10, color='red')


plt.subplot(1,2,2)
plt.plot(data['Protein'], data['Carb. (g)'], marker='o', color='orange', markersize='0.2', markerfacecolor='red', linestyle='-')
plt.xlabel('Protein')
plt.ylabel('Carb. (g)')
plt.title(' Protein (g) vs Carb. (g) Line Plot', fontsize=10, color='red')
```

```python
plt.subplot(1,2,2)
plt.bar(x=df['Calories'].unique(), height=df['Calories'].value_counts().to_list(), color=['#0f5a3e', '#f8b195', '#f67280'])
plt.xlabel('Calories')
plt.ylabel('Count')
plt.title('Count of Calories', fontsize=20, color='red')
# plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(data['Fiber (g)'], data['Protein'], marker='o', color='blue', markersize='3.0', markerfacecolor='blue', linestyle='solid')
plt.fill_between(data['Fiber (g)'], data['Protein'], color='#800080') # Removed extra ')' from 'Protein)'
plt.xlabel('Fiber (g)')
plt.ylabel('Protein')
plt.title('Protein vs Fiber (g) Line Plot', fontsize=20, color='red')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=df[['Calories', 'Fat (g)', 'Protein', 'Fiber (g)', 'Carb. (g)']], palette=['#FF00FF', '#A52A2A', '#0000FF', '#FFC0CB','#FFA500'])
plt.xlabel(' Categories')
plt.ylabel(' Count')
plt.title('Count of Categories (Boxplot)', fontsize=20, color='Red')
plt.show()
```

```python
# Wordcloud
from wordcloud import WordCloud

wordcloud = WordCloud().generate(all_words)
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
plt.figure(figsize=(10,10))
wordcloud = WordCloud(width=500, height=400, background_color='black').generate(all_words)
plt.imshow(wordcloud, interpolation='bilinear')
plt.title('WordCloud for Nutrition Drinks', fontsize=20, color='red')
plt.axis('off')
plt.show()
```

```python

sns.jointplot(data=df, x='Calories', y='Fat (g)', kind='hex', color='#FF00FF')
plt.xlabel('Calories')
plt.ylabel('Fat (g)')
plt.xticks(rotation=90)
plt.yticks(rotation=90)

plt.title('Calories Joint Plot', fontsize=20, color='red')
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.cluster.hierarchy import fcluster
from scipy.spatial.distance import pdist
```
