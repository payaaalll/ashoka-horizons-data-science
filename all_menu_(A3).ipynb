```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly as py
data = pd.read_csv('all-menu.csv')
```

```python
plt.figure(figsize=(15,15))
plt.plot(data['Calories'], data[' Total Fat (g)'], data[' Protein (g) '], data[' Sugars (g)'])

plt.title(' Calories vs Fat (g) vs Protein vs (g) Carb. (g) Line Plot', fontsize=20, color='red')

```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 7))
axs[0].plot(data['Calories'], color='blue')
axs[0].set_xlabel('Index')
axs[0].set_ylabel('Calories')
axs[0].set_title('Calories Line Plot', fontsize=16, color='red')
axs[1].plot(data[' Protein (g) '], color='green')
axs[1].set_xlabel('Index')
axs[1].set_ylabel('Protein (g)')
axs[1].set_title('Protein (g) Line Plot', fontsize=16, color='red')
```

```python
plt.subplot(1,2,1)
plt.bar(x=df['Calories'].unique(), height=df['Calories'].value_counts().to_list(), color=['#0f5a3e', '#f8b195', '#f67280'])
plt.xlabel('Calories')
plt.ylabel('Count')
plt.title('Count of Calories', fontsize=20, color='green')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(data['Calories'], data[' Total Fat (g)'], marker='o', color='blue', markersize='3.0', markerfacecolor='blue', linestyle='solid')
plt.fill_between(data['Calories'], data[' Total Fat (g)'], color='#FFFF00')
plt.xlabel('Calories')
plt.ylabel(' Total Fat (g)')
plt.title('Calories vs Total Fat (g) Line Plot', fontsize=20, color='green')
plt.show()
```

```python
plt.figure(figsize=(15,5))
sns.boxplot(data=df[['Calories', ' Total Fat (g)', ' Protein (g) ', ' Sugars (g)']], palette=['#FF00FF', '#A52A2A', '#0000FF', '#FFC0CB','#FFA500'])
plt.xlabel(' Categories')
plt.ylabel(' Count')
plt.title('Count of Categories (Boxplot)', fontsize=20, color='Red')
plt.show()
```

```python
fig, axs = plt.subplots(1, 2, figsize=(15, 7))
sns.boxplot(y='Calories', data=df, ax=axs[0], palette=['#FF00FF'])
axs[0].set_ylabel('Calories')
axs[0].set_xlabel('Categories')
axs[0].set_title('Count of Calories (Boxplot)', fontsize=16, color='red')

sns.boxplot(y=' Total Fat (g)', data=df, ax=axs[1], palette=['#A52A2A'])
axs[1].set_ylabel('Total Fat (g)')
axs[1].set_xlabel('Categories')
axs[1].set_title('Count of Total Fat (g) (Boxplot)', fontsize=16, color='red')
```
